```python
##QNO:2:
url="https://www.imdb.com/chart/top/?sort=ir,desc&mode=simple&page=1"
page=requests.get(url)
page
```

```python
page.content
soup=BeautifulSoup(page.content)
scraped_movies=soup.find_all('td',class_='titleColumn')
##scraped_movies

```

```python

movies=[]
count=0
if count<=99:
    for movie in scraped_movies:
        movie=movie.get_text().replace('\n',"")
        movie=movie.strip("")
        movie=movie.replace('', "")
        movies.append(movie)
        count+=1
        if count>=100:
            break
               
##movies

scraped_ratings=soup.find_all('td',class_='ratingColumn imdbRating')
##scraped_ratings
ratings=[]
count=0
if count<=99:
    for rating in scraped_ratings:
        rating=rating.get_text().replace('\n',"")
        ratings.append(rating)
        count+=1
        if count>=100:
            break
    
##ratings

import pandas as pd
data=pd.DataFrame(index=None)
data['MovieNames']=movies
data['Ratings']=ratings
data

    
```

```python
##QNO:4
page=requests.get('https://presidentofindia.nic.in/former-presidents.htm')
page
```

```python
soup=BeautifulSoup(page.content)
#print(soup.prettify())
```

```python
PresidentNames=[]
heading_tags = ['h3']
for tags in soup.find_all(heading_tags):
    PresidentNames.append(tags.get_text())  
print(PresidentNames)
    
office=[]
for i in soup.find_all("div",class_="presidentListing"):
    office.append(i.find_all("p")[0].text.split(":")[1])
print(office)

        
    
```

```python
##QNo:5 a) ODI Teams
url="https://www.icc-cricket.com/rankings/mens/team-rankings/odi"
page=requests.get(url)
page
soup=BeautifulSoup(page.content)
##print(soup.prettify())
```

```python

Matches=[]
Points=[]
TeamNames=[]
Rating=[]

count=0
table = soup.find_all('table')[0].find_all('tr')
for row in table:    
    # Find all data for each column
    columns = row.find_all('td')
    if(columns!= []):
        
        TeamNames.append(columns[1].text.strip().replace("\n",""))
        Matches.append(columns[2].text.strip())
        Points.append(columns[3].text.strip())
        Rating.append(columns[4].text.strip())
        count+=1
        if count>8:
            break 


import pandas as pd
data=pd.DataFrame()
data['Team']=TeamNames
data['Matches']=Matches
data['Points']=Points
data['Rating']=Rating
data.style.hide_index()
                                        

```

```python
## QNO 5(b) BATSMAN RANKINGS ODI:
url="https://www.icc-cricket.com/rankings/mens/player-rankings/odi"
page=requests.get(url)
page
soup=BeautifulSoup(page.content)
##print(soup.prettify())
```

```python

##QNO:7
url="https://www.cnbc.com/world/?region=world"
page=requests.get(url)

page
```

```python
soup=BeautifulSoup(page.content,'html.parser')
#print(soup.prettify())
```

```python
Timestamps=[]
for i in soup.find_all("span",class_="LatestNews-wrapper"):
    Timestamps.append(i.find("time").text)
##print(Timestamps)

Headlines=[]
for news in soup.find_all("div",class_="LatestNews-headlineWrapper"):
    Headlines.append(news.a.text)
##print(Headlines)

Hyperlinks=[]
for i in soup.find_all("a",href=True,class_="LatestNews-headline"):
    Hyperlinks.append(i['href'])
##print(Hyperlinks)

import pandas as pd
data=pd.DataFrame()
data['Headlines']=Headlines
data['Hyperlinks']=Hyperlinks
data['Timestamps']=Timestamps

data.style.hide_index()
```

```python
data.to_csv(r'C:/Users/hp/Latestnewscnbc.csv')
```

```python
##QNO:9
url="https://www.dineout.co.in/delhi-restaurants/buffet-special"
page=requests.get(url)

page
```

```python
soup=BeautifulSoup(page.content)
##print(soup.prettify())
```

```python
Restaurants=[]
for i in soup.find_all("a",class_="restnt-name ellipsis"):
    Restaurants.append(i.text)
print(Restaurants)
```

```python
Location=[]
for i in soup.find_all("div",class_="restnt-loc ellipsis"):
    Location.append(i.text)
print(Location)

```

```python
Cuisine=[]
for i in soup.find_all('span',class_="double-line-ellipsis"):
    Cuisine.append(i.text.split()[6])
print(Cuisine)

```

```python
Rating=[]
for i in soup.find_all('div',class_="restnt-rating rating-4"):
    Rating.append(i.text)
print(Rating)

```

```python
Images=[]
for i in soup.find_all('img',class_="no-img"):
    Images.append(i['data-src'])
##print(Images)
```

```python
import pandas as pd
data=pd.DataFrame()
data['Restaurants']=Restaurants
data['Location']=Location
data['Cuisine']=Cuisine
data['Rating']=Rating
data['Images URL']=Images

data
```

```python
data.to_csv(r'C:/Users/hp/Restaurants.csv')
```

```python
##QNO:10
url="https://scholar.google.com/citations?view_op=top_venues&hl=en"
page=requests.get(url)

page
```

```python
soup=BeautifulSoup(page.content)
##print(soup.prettify())
```

```python
Rank=[]
for i in soup.find_all('td',class_="gsc_mvt_p"):
    Rank.append(i.text)
print(Rank)
```

```python
Publication=[]
for i in soup.find_all('td',class_="gsc_mvt_t"):
    Publication.append(i.text)
##print(Publication)
```

```python
h5index=[]
for i in soup.find_all('a',class_="gs_ibl gsc_mp_anchor"):
    h5index.append(i.text)
print(h5index)
```

```python
h5median=[]
for i in soup.find_all('span',class_="gs_ibl gsc_mp_anchor"):
    h5median.append(i.text)
print(h5median)
```

```python
import pandas as pd
data=pd.DataFrame()
data['Rank']=Rank
data['Publication']=Publication
data['h5index']=h5index
data['h5median']=h5median

data.style.hide_index()
```
